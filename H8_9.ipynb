```python
kata = input('masukkan kata: ')
#kata = 'BaT'
#huruf = [str(huruf) for huruf in kata]

print('kata\t= ', kata)
#print('huruf= ', huruf)


for i in range(len(kata)):
    print('\nhuruf\t= ', kata[i])
    desimal1 = ord(kata[i])
    print('desimal\t= ', desimal1)
    
    biner = ''
    oktal = ''
    hexa = ''

    #biner
    desimal=desimal1
    while desimal > 0:
        sisa = desimal % 2
        biner = str(sisa) + biner
        desimal //= 2
    print('biner\t= ', biner)

    #oktal
    desimal=desimal1
    while desimal > 0:
        sisa = desimal % 8
        oktal = str(sisa) + oktal
        desimal //= 8
    print('oktal\t= ', oktal)

    #hexa
    desimal=desimal1
    hex_chars = "0123456789ABCDEF"
    while desimal > 0:
         sisa = desimal % 16
         hexa = hex_chars[sisa] + hexa
         desimal //= 16
    print('hexa\t= ', hexa)
```

```python
#nama file
path='handling.txt'

#membuat file
try:
    with open(path, 'x'):
        f_op = open(path, 'a')
        f_op.write(f'''
{'*'*42}\n
{"DATA PESERTA":^42}\n
{'*'*42}
''')
        f_op.close()
    print(f"The file '{path}' has been created successfully.")
except FileExistsError:
    print(' ')

#input 
nama = input('Nama\t: ')
ruang = input('Ruang\t:')
skema = input('Skema\t:')

#tulis ke file
teks1 = f'''{"="*42}
Nama\t: {nama}
Ruang\t: {ruang}
Skema\t: {skema}
{"-"*42}'''
teks2 = f"\n|{"huruf":^5}|{'biner':^9}|{'oktal':^7}|{'desimal':^9}|{'hexa':^6}|"
f_op = open(path, 'a')
f_op.write(teks1)
f_op.write(teks2)
f_op.close()



for i in range(len(nama)):
    desimal1 = ord(nama[i])
    biner = ''
    oktal = ''
    hexa = ''

    #biner
    desimal=desimal1
    while desimal > 0:
        sisa = desimal % 2
        biner = str(sisa) + biner
        desimal //= 2
    
    #oktal
    desimal=desimal1
    while desimal > 0:
        sisa = desimal % 8
        oktal = str(sisa) + oktal
        desimal //= 8
   

    #hexa
    desimal=desimal1
    hex_chars = "0123456789ABCDEF"
    while desimal > 0:
         sisa = desimal % 16
         hexa = hex_chars[sisa] + hexa
         desimal //= 16
    
    hasil = "\n|{:^5}|{:>9}|{:>7}|{:>9}|{:>6}|".format(nama[i],biner,oktal,desimal1,hexa)
    f_op = open(path, 'a')
    f_op.write(hasil)
    print(hasil) 

f_op.write('\n\n')
f_op.close()      

print("sukses, cek file handling.txt")
```

```python
import pandas as pd
import mysql.connector


mydb = mysql.connector.connect(
host="127.0.0.1",
user="root",
password="root",
database="sakila",
)

nama = input('Input Name: ')

query = f'''
SELECT film.title, category.name AS kategori, payment.amount
FROM customer
INNER JOIN rental ON customer.customer_id = rental.customer_id
INNER JOIN payment ON rental.rental_id = payment.rental_id
INNER JOIN inventory ON rental.inventory_id = inventory.inventory_id
INNER JOIN film ON inventory.film_id = film.film_id
INNER JOIN film_category ON film.film_id = film_category.film_id
INNER JOIN category ON film_category.category_id = category.category_id
WHERE customer.first_name LIKE '%{nama}%'
'''

df = pd.read_sql_query(query, mydb)
df
```

```python
import pandas as pd
import mysql.connector


mydb = mysql.connector.connect(
host="127.0.0.1",
user="root",
password="root",
database="sakila",
)

category = input('Kategori: ')

query = f'''
SELECT CONCAT (customer.first_name," ",customer.last_name) as nama, payment.amount as amount
FROM customer
INNER JOIN rental ON customer.customer_id = rental.customer_id
INNER JOIN payment ON rental.rental_id = payment.rental_id
INNER JOIN inventory ON rental.inventory_id = inventory.inventory_id
INNER JOIN film ON inventory.film_id = film.film_id
INNER JOIN film_category ON film.film_id = film_category.film_id
INNER JOIN category ON film_category.category_id = category.category_id
WHERE category.name LIKE '%{category}%'
'''

df = pd.read_sql_query(query, mydb)
df
```
